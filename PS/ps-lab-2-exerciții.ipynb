## Exercițiul 1

Fie semnalele continue $x(t) = \cos (520 \pi t + \pi/3)$, $y(t) = \cos (280 \pi t - \pi/3)$ și $z(t) = \cos (120 \pi t + \pi/3)$.

- În Python, simulați axa reală de timp printr-un șir de numere suficient de apropiate, spre exemplu $[0:0.0005:0.03]$ (adică intervalul $[0, 0.03]$, discretizat cu un pas de $0.0005$).

```python
start = 0
Ts = 0.0005
stop = 0.03

t = np.arange(start, stop + Ts, Ts)

plt.xlabel('t [s]')
plt.ylabel('Amplitudine')
plt.title('Axa reala de timp')
plt.grid(True)
plt.stem(t, np.zeros_like(t))
```

- Construiți semnalele $x(t)$, $y(t)$ și $z(t)$ și afișați-le grafic, în câte un subplot.

```python
x = lambda t: np.cos(520 * np.pi * t + np.pi / 3)
y = lambda t: np.cos(280 * np.pi * t - np.pi / 3)
z = lambda t: np.cos(120 * np.pi * t + np.pi / 3)

fig, axs = plt.subplots(3, figsize = (10, 8))

plt.subplots_adjust(hspace = 0.6)

axs[0].set_title('x(t)')
axs[0].set_xlabel('Timp (s)')
axs[0].set_ylabel('Amplitudine')
axs[0].grid(True)
axs[0].plot(t, x(t))

axs[1].set_title('y(t)')
axs[1].set_xlabel('Timp (s)')
axs[1].set_ylabel('Amplitudine')
axs[1].grid(True)
axs[1].plot(t, y(t))

axs[2].set_title('z(t)')
axs[2].set_xlabel('Timp (s)')
axs[2].set_ylabel('Amplitudine')
axs[2].grid(True)
axs[2].plot(t, z(t))
```

- Eșantionați semnalele cu o frecvență de 200 Hz pentru a obține $x[n]$, $y[n]$ și $z[n]$ și afișați-le grafic, în câte un subplot.

```python
fs = 200
Ts = 1 / fs

t_n = np.arange(start, stop + Ts, Ts)

fig, axs = plt.subplots(3, figsize = (10, 8))

plt.subplots_adjust(hspace = 0.6)

axs[0].set_title('x(t_n)')
axs[0].set_xlabel('Timp (s)')
axs[0].set_ylabel('Amplitudine')
axs[0].grid(True)
axs[0].stem(t_n, x(t_n))

axs[1].set_title('y(t_n)')
axs[1].set_xlabel('Timp (s)')
axs[1].set_ylabel('Amplitudine')
axs[1].grid(True)
axs[1].stem(t_n, y(t_n))

axs[2].set_title('z(t_n)')
axs[2].set_xlabel('Timp (s)')
axs[2].set_ylabel('Amplitudine')
axs[2].grid(True)
axs[2].stem(t_n, z(t_n))
```

## Exercițiul 2

Generați semnale descrise mai jos și afișați-le grafic, fiecare într-un plot. Nu aveți voie să folosiți funcțiile predefinite din `scipy.signal`.

- Un semnal **sinusoidal** de frecvență 400 Hz, care să conțină 1600 de eșantioane.

```python
f = 400

start = 0
N = 1600
fs = 500 * f
Ts = 1 / fs
stop = Ts * N

t = np.arange(start, stop, Ts)

x = lambda t : np.sin(2 * np.pi * f * t)

plt.xlabel('t [s]')
plt.ylabel('Amplitudine')
plt.title('x(t)')
plt.grid(True)
plt.plot(t, x(t))
```

- Un semnal **sinusoidal** de frecvență 800 Hz, care să dureze 3 secunde.

```python
f = 800

start = 0
fs = 0.14 * f
Ts = 1 / fs
stop = 3

t = np.arange(start, stop, Ts)

x = lambda t : np.sin(2 * np.pi * f * t)

plt.xlabel('t [s]')
plt.ylabel('Amplitudine')
plt.title('x(t)')
plt.grid(True)
plt.plot(t, x(t))
```

- Un semnal de tip **sawtooth** de frecvență 240 Hz (puteți folosi funcțiile `numpy.floor` sau `numpy.mod`).

```python
f = 240

start = 0
fs = 0.14 * f
Ts = 1 / fs
stop = 3

t = np.arange(start, stop, Ts)

x = lambda t : 2 * (f * t - np.floor(f * t + 0.5))

plt.xlabel('t [s]')
plt.ylabel('Amplitudine')
plt.title('x(t)')
plt.grid(True)
plt.plot(t, x(t))
```

- Un semnal de tip **square** de frecvență 300 Hz (puteți folosi funcția `numpy.sign`).

```python
f = 300

start = 0
fs = 0.14 * f
Ts = 1 / fs
stop = 3

t = np.arange(start, stop, Ts)

x = lambda t : np.sign(np.sin(2 * np.pi * f * t))

plt.xlabel('t [s]')
plt.ylabel('Amplitudine')
plt.title('x(t)')
plt.grid(True)
plt.plot(t, x(t))
```

- Un semnal 2D **la alegerea voastră**. Creați un `numpy.array` de dimensiune 128x128 și inițializați-l folosind o procedură creată de voi.
Utilizați, spre exemplu, funcțiile `numpy.zeros()` și `numpy.ones()`.

```python
size = 128

I = np.zeros((size, size))

centerX, centerY = size // 2, size // 2

radiusMax = size // 2

for radius in range(0, radiusMax, size // 8):
    for i in range(size):
        for j in range(size):
            if (i - centerX)**2 + (j - centerY)**2 <= radius**2:
                I[i, j] = min(i, j)

plt.imshow(I, cmap = 'gray')
```

## Exercițiul 3

Un semnal este înregistrat cu o frecvență de eșantionare de 2000 Hz.

- Care este intervalul de timp între două eșantioane?

- Dacă un eșantion este memorat pe 4 biți, câți bytes vor ocupa 1 oră de achiziție?

## Exercitiul 3.1
$$ T_s = \frac{1}{f_s} $$
$$ f_s = 2000 Hz = 2 * 10^3 \frac{esantioane}{s} $$
$$ T_s = \frac{1}{2 * 10^3 \frac{1}{s}} = 0.5 * 10^{-3} s = 0.5 ms = 0.0005 \frac{s}{esantion} $$

## Exercitiul 3.2
$$ t_s = 1 h = 3600 s = 3.6 * 10^3 s $$
$$ esantioane = f_s * t_s = 2 * 10^3 \frac{esantioane}{s} * 3.6 * 10^3 s = 7.2 * 10^6 esantioane $$
$$ n = 4 \frac{biti}{esantion} = 4 \frac{biti}{esantion} * \frac{1}{8} \frac{bytes}{biti} = 0.5 \frac{bytes}{esantion} $$
$$ bytes = esantioane * n = 7.2 * 10^6 esantioane * 0.5 \frac{bytes}{esantion} = 3.6 * 10^6 bytes = 3.6 MB $$

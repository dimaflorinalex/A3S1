## Exercițiul 4

Utilizați funcția `np.fft.rfft(x)` pentru a calcula **transformata Fourier** a semnalului și afișați grafic **modulul transformatei**.

Deoarece valorile pe care le veți calcula sunt în Hz, este important să definiți corect frecvența de eșantionare (astfel încât valorile de frecvențe pe care le obțineți utilizând ultima secvență de cod din ghidul Python din îndrumar să aibă o interpretare corectă din punct de vedere fizic).

```python
data = np.genfromtxt("Train.csv", delimiter = ",", skip_header=1)
x = data[:, 2]
fs = 0.000277
ts = 3600
N = 18288
fs = 1 / ts
X = np.fft.rfft(x)
X = np.abs(X/N)
freqs = np.fft.rfftfreq(len(x), d=ts)
freqs = fs * np.linspace(0, N//2, N//2 + 1) / N
amps = np.abs(X)
plt.figure(figsize=(12, 6))
plt.plot(freqs, amps)
plt.grid()
plt.show()
```

## Exercițiul 5

Prezintă acest semnal o **componentă continuă**? Dacă da, eliminați-o și afișați semnalul rezultat. Dacă nu, specificați cum ați determinat acest fapt.

```python
mean_x = np.mean(x)

if mean_x != 0:
    print(f"Componenta continua: {mean_x:.2f}")
    x_centered = x - mean_x
    mean_xc = np.mean(x_centered)
    plt.figure(figsize=(12, 6))

    plt.subplot(2, 1, 1)
    plt.plot(x, label="Semnal original")
    plt.axhline(y=mean_x, color='r', linestyle='--', label=f"Media ({mean_x:.2f})")
    plt.legend()
    plt.grid()

    plt.subplot(2, 1, 2)
    plt.plot(x_centered, label="Semnal fara componenta continua", color='g')
    plt.axhline(y=mean_xc, color='r', linestyle='--', label=f"Media ({mean_xc:.2f})")
    plt.legend()
    plt.grid()

    plt.tight_layout()
    plt.show()
else:
    print("Semnalul nu are componenta continua")
```

## Exercițiul 6

Care sunt frecvențele principale conținute în semnal, așa cum apar ele în transformata Fourier? Mai exact, determinați **primele 4 cele mai mari valori** ale modulului transformatei și specificați căror frecvențe (în Hz) le corespund. Căror fenomene periodice din semnal se asociază fiecare?

```python
amps[0] = 0
top_idx = np.argsort(amps)[-4:][::-1]
top_freqs = freqs[top_idx]
top_amps = amps[top_idx]

for i, (freq, amp) in enumerate(zip(top_freqs, top_amps)):
    period = 1 / freq if freq != 0 else float("inf")
    freq = freq * 10**6
    print(f"Frecventa #{i+1}: {freq:.3f} μHz, Magnitudine: {amp:.2f}, Perioada: {period}s")
```

## Exercițiul 7

Începând de la un eșantion ales de voi (după cel al 1000-lea), reprezentați, pe un grafic separat, **o lună de trafic**. Alegeți eșantionul de start astfel încât reprezentarea să înceapă într-o **zi de luni**.

```python
import pandas as pd 

data = pd.read_csv("Train.csv")
data["Datetime"] = pd.to_datetime(data["Datetime"], format="%d-%m-%Y %H:%M")
subset = data.iloc[13000:]
start_idx = subset[subset["Datetime"].dt.weekday == 0].index[0]
end_idx = start_idx + 720
month_data = data.loc[start_idx:end_idx]

plt.figure(figsize=(12, 6))
plt.plot(month_data["Datetime"], month_data["Count"], label="Trafic")
plt.xlabel("T")
plt.ylabel("Nr de masini")
plt.grid()
plt.legend()
plt.show()
```

## Exercițiul 1

 Scrieți câte o funcție prin care să construiți o **fereastră    dreptunghiulară**, respectiv o **fereastră de tip Hanning**. Funcțiile primesc ca parametru **lungimea ferestrei**.

 Generați și afișați o sinusoidă cu frecvență $f = 100 \, \text{Hz}$, amplitudine unitară și fază nulă, apoi treceți-o prin cele două tipuri de ferestre (pentru lungimea $N_w = 200$) și afișați grafic rezultatul.

```python
def rectangular_window(N):
    return np.ones(N)

def hanning_window(N):
    return np.hanning(N)

f = 100
fs = 1000
t = np.arange(0, 1, 1/fs)
signal = np.sin(2 * np.pi * f * t)

N_w = 200
window_rect = rectangular_window(N_w)
window_hann = hanning_window(N_w)

signal_segment = signal[:N_w]

signal_rect = signal_segment * window_rect
signal_hann = signal_segment * window_hann

plt.figure(figsize=(12, 6))

plt.subplot(3, 1, 1)
plt.plot(signal_segment, label="Semnal initial")
plt.title("Semnal initial")
plt.grid()
plt.legend()

plt.subplot(3, 1, 2)
plt.plot(signal_rect, label="Semnal prin fereastra dreptunghiulara")
plt.title("Semnal prin fereastra dreptunghiulara")
plt.grid()
plt.legend()

plt.subplot(3, 1, 3)
plt.plot(signal_hann, label="Semnal prin fereastra Hanning")
plt.title("Semnal prin fereastra Hanning")
plt.grid()
plt.legend()

plt.tight_layout()
plt.show()
```

## Exercițiul 2

Fișierul `traffic-one-week.csv` conține date de trafic colectate pe o perioadă de o săptămână. Perioada de eșantionare este de o oră, iar valorile măsurate reprezintă numărul de vehicule care trec printr-o intersecție.

### Subpunctul a)

Încărcați datele în memorie și selectați din semnalul dat o porțiune corespunzătoare pentru **3 zile**, `x`, pe care veți lucra în continuare.

```python
signal = np.genfromtxt('ps-lab-9-traffic-one-week.csv')
x = signal[:72]
```

### Subpunctul b)

Utilizați funcția `np.convolve(x, np.ones(w), 'valid') / w` pentru a realiza un filtru de tip medie alunecătoare și neteziți semnalul obținut anterior. Setați dimensiuni diferite ale ferestrei (variabila `w` în codul de mai sus), spre exemplu $5$, $9$, $13$, $17$. Afișați grafic semnalele obținute.

```python
window_sizes = [5, 9, 13, 17]

smoothed_signals = [np.convolve(x, np.ones(w), 'valid')/w for w in window_sizes]

plt.figure(figsize=(10, 6))
plt.plot(x, label="Semnal initial")

for w, smoothed in zip(window_sizes, smoothed_signals):
    plt.plot(range(w - 1, 72), smoothed, label=f"Medie alunecatoare (w={w})")

plt.title("Filtrare cu medie alunecatoare")
plt.xlabel("Ora")
plt.ylabel("Numar vehicule")
plt.legend()
plt.grid(True)

plt.show()
```

### Subpunctul c)

Dorinți să filtrați **zgomotul** (frecvențe înalte) din semnalul cu date de trafic. Alegeți o **frecvență de tăiere** pentru un filtru trece-jos pe care îl veți crea în subpunctul următor. Argumentați alegerea făcută.

**Indicație:** transformata Fourier a semnalului, calculată în laboratorul anterior, ar putea să vă ajute cu luarea acestei decizii.

Pe lângă valoarea frecvenței în Herzi, determinați și valoarea ei normalizată între 0 și 1, unde 1 reprezintă frecvența Nyquist pentru semnalul de trafic.

```python
X = np.fft.fft(x)
freqs = np.fft.fftfreq(len(x))

X_magnitude = np.abs(X)

plt.figure(figsize=(10, 5))
plt.plot(freqs[:len(freqs)//2], X_magnitude[:len(X_magnitude)//2], label="Spectru")
plt.title("Frecventele semnalului")
plt.xlabel("Frecventa normalizata")
plt.ylabel("Amplitudine")
plt.grid(True)
plt.show()

f_taiere = 0.1
f_taiere_normalizata = f_taiere * 2

print(f"Frecventa de taiere aleasa: {f_taiere} (normalizata: {f_taiere_normalizata})")
```

### Subpunctul d)

Utilizați funcția `scipy.signal.butter`, respectiv funcția `scipy.signal.cheby1`, pentru a **proiecta** filtrele Butterworth și Chebyshev de ordin 5, cu frecvența de tăiere $W_n$ (valoare normalizată), stabilită la subpunctul anterior.
        
Pentru început, setați atenuarea ondulațiilor la `rp = 5` (dB), urmând ca apoi să încercați și alte valori.

```python
from scipy.signal import butter, cheby1, lfilter

f_taiere_normalizata = 0.1
ordine = 5
rp = 5

def apply_filter(filtru_b, filtru_a, semnal):
    return lfilter(filtru_b, filtru_a, semnal)

b_butter, a_butter = butter(ordine, f_taiere_normalizata, btype='low', analog=False)
semnal_butter = apply_filter(b_butter, a_butter, x)

b_cheby1, a_cheby1 = cheby1(ordine, rp, f_taiere_normalizata, btype='low', analog=False)
semnal_cheby1 = apply_filter(b_cheby1, a_cheby1, x)

# Afișarea rezultatelor
plt.figure(figsize=(12, 6))
plt.plot(x, label='Semnal initial')
plt.plot(semnal_butter, label='Filtru Butterworth')
plt.plot(semnal_cheby1, label='Filtru Chebyshev')
plt.title('Filtrarea semnalului de trafic cu filtre Butterworth și Chebyshev')
plt.xlabel('Ora')
plt.ylabel('Numar vehicule')
plt.legend()
plt.grid(True)

plt.show()
```

### Subpunctul e)

**Filtrați** datele de trafic cu cele 2 filtre și afișați semnalele filtrate împreună cu datele brute. Ce filtru alegeți dintre cele două și de ce?

```python
from scipy.signal import butter, cheby1, lfilter

f_taiere_normalizata = 0.1
ordine = 5
rp = 5

def apply_filter(filtru_b, filtru_a, semnal):
    return lfilter(filtru_b, filtru_a, semnal)

b_butter, a_butter = butter(ordine, f_taiere_normalizata, btype='low', analog=False)
semnal_butter = apply_filter(b_butter, a_butter, x)

b_cheby1, a_cheby1 = cheby1(ordine, rp, f_taiere_normalizata, btype='low', analog=False)
semnal_cheby1 = apply_filter(b_cheby1, a_cheby1, x)

# Afișarea rezultatelor
plt.figure(figsize=(12, 6))
plt.plot(x, label='Semnal initial')
plt.plot(semnal_butter, label='Filtru Butterworth')
plt.plot(semnal_cheby1, label='Filtru Chebyshev')
plt.title('Filtrarea semnalului de trafic cu filtre Butterworth și Chebyshev')
plt.xlabel('Ora')
plt.ylabel('Numar vehicule')
plt.legend()
plt.grid(True)

plt.show()

print('Filtrul Butterworth: preferat pentru o reprezentare neteda și lipsita de ondulatii, fiind util în analiza tendintei generale a datelor de trafic.')
print('Chebyshev: Util unde este importanta o tranziție mai abrupta intre banda de trecere si banda de oprire, dar poate introduce ondulatii care distorsioneaza interpretarea vizuala.')
print('In acest caz, filtrul Butterworth este mai potrivit pentru analizarea traficului, oferind o estimare mai stabila si mai usor de interpretat.')
```

### Subpunctul f)

Reproiectați filtrele alegând atât un ordin mai mic, cât și unul mai mare. De asemenea, reproiectați filtrul Chebyshev cu alte valori ale `rp` și observați efectul.

Stabiliți **valorile optime** ale parametrilor pentru a vă atinge scopul (cel de a filtra zgomotul din semnal).

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import butter, cheby1, lfilter

# Încărcarea datelor
data = np.genfromtxt("ps-lab-9-traffic-one-week.csv")
x = data[:72]

f_taiere_normalizata = 0.1
ordine_valori = [3, 5, 7]
rp_valori = [1, 5, 10]

def apply_filter(filtru_b, filtru_a, semnal):
    return lfilter(filtru_b, filtru_a, semnal)

plt.figure(figsize=(14, 8))
plt.plot(x, label='Semnal initial')

for ordine in ordine_valori:
    b_butter, a_butter = butter(ordine, f_taiere_normalizata, btype='low', analog=False)
    semnal_butter = apply_filter(b_butter, a_butter, x)
    plt.plot(semnal_butter, label=f'Butterworth, ordin={ordine}')

plt.title('Compararea filtrelor Butterworth cu ordine diferite')
plt.xlabel('Ora')
plt.ylabel('Numar vehicule')
plt.legend()
plt.grid(True)
plt.show()

plt.figure(figsize=(14, 8))
plt.plot(x, label='Semnal initial')

for rp in rp_valori:
    b_cheby1, a_cheby1 = cheby1(5, rp, f_taiere_normalizata, btype='low', analog=False)
    semnal_cheby1 = apply_filter(b_cheby1, a_cheby1, x)
    plt.plot(semnal_cheby1, label=f'Chebyshev, rp={rp} dB')

plt.title('Compararea filtrelor Chebyshev cu valori rp diferite')
plt.xlabel('Ora')
plt.ylabel('Numar vehicule')
plt.legend()
plt.grid(True)
plt.show()

# Observatii:
# 1. Un ordin mai mare al filtrului Butterworth => o netezire mai puternica, dar poate introduce delay.
# 2. Pentru filtrul Chebyshev: o valoare mai mare a lui rp => mai multe ondulatii in banda de trecere.
# Valori optime sugerate: 
# 1. Ordinul 5 pentru Butterworth
# 2. rp = 5 dB pentru Chebyshev (tranzitie echilibrata fara multe ondulatii)
```
